## История

*LaTeX: TeX — название системы верстки, разработанной Дональдом Кнутом, Lamport — в честь Лесли Лампорта, создателя LaTeX.*

В 1977 году Дональд Кнут получил гранки переиздания его книги "Искусство программирования", и был настолько разочарован, что решил создать свою собственную систему набора. Название TeX было выбрано как игра слов, основанная на латинском слове "techne" (техника, искусство), или как сокращение от "tech" (технология) и "X" — символа, часто используемого в математике и научных дисциплинах. И в начале 1980-х годов Лесли Лэмпорту понадобилось написать макросы TeX для собственного использования, и он подумал, что, приложив немного дополнительных усилий, он сможет сделать общий пакет, пригодный для использования другими. Так появился LaTeX.

*SAGE: System of Arithmetic Geometry Experimentation*

В 2005 году американский учёный Уильям Штейн (William Stein) осознал необходимость создания прогаммного обеспечения для более эффективного с вычислительной точки зрения решения задач из области теории чисел. Позже мировое сообщество распространило его на другие разделы математики.

## Основные операции с числами

- сложение: ```a+b```
- вычитание: ```a-b```
- умножение: ```a*b```
- деление: ```a/b```
- возведение в степень: ```a^b``` или ```a**b```
- извлечение квадратного корня: ```sqrt(a)```
- извлечение корня степени $n$: ```a^(1/n)```
- целая часть при делении: ```a//b```
- остаток при делении: ```a%b```
- получение пары (целая часть, остаток): ```divmod(a,b)```
- вычисление факториала $n!$: ```factorial(n)```
- вычисление биномиальных коэффициентов $C_n^k = \binom{n}{k}$: ```binomial(n,k)```
- округление "снизу" - до ближайшего целого, не превосходящего заданное число $n$: ```floor(n)```

Часто встречающиеся математические функции:
- экспонента и логарифм ```exp, log```
- логарифм по основанию $a$ от $x$ или $log_a x$: ```log(x, a)```

- арксинус, арккосинус, арктангенс: ```arcsin, arccos, arctan```

Полезно использовать функцию ```bool(expression)```, которая возвращает ```True```, если выражение верно, и ```False``` в противоположном случае.
