## Параграфы

Для перехода к новому параграфу можно воспользоваться двумя способами:
- поставить в конце строки два символа пробела
- вставить пустую строку между концом предыдущего параграфа и началом следующего
- добавить в конце строки последовательность символов `<br>` из HTML
- некоторые редакторы поддерживают знак `\`  в конце строки, но такой метод не рекомендуют во избежание ошибок

Отметим, что конечные пробелы сложно заметить в редакторе, и многие случайно или намеренно добавляют два пробела после каждого предложения. Поэтому лучшими вариантами являются добавление пустой строки или символов `<br>`.

## Списки: нумерованные, вложенные


## Форматирование текста

В особый класс выделены служебные символы, отвечающие за создание заголовков, форматирование текста, которые невозможно при необходимости просто так отобразить на экране, но для этого придуман способ экранирования, то есть написание наклонной черты влево `\` перед символом:

|Символ |	Название на английском| Написание с экранированием |
|:------|:------|:------|
|`\`	| backslash | `\\`|
|`  	|backtick (see also escaping backticks in code) | ``` \` ```|
|`*`	|asterisk |`\*`|
|`_`	|underscore | `\_` |
|`{ }`	|curly braces | `\{ \}`|
|`[ ]`	|brackets |`\[ \]` |
|`< >`	|angle brackets |`\< \>` |
|`( )`	|parentheses |`\( \)` |
|`#`	|pound sign  |`\#` |
|`+`	|plus sign   | `\+`|
|`-`	|minus sign (hyphen) |`\-` |
|`.`	|dot   | `\.`|
|`!`	|exclamation mark |`\!` |
|`\|`    |pipe (see also escaping pipe in tables) | \\\| |

Для того, чтобы сделать форматирование слов, необходимо окружить текст специальными символами:
- *наклонный шрифт* `*Italic*  or _Italic_`
- **полужирный шрифт** `**Bold** or __Bold__`
- ***полужирный наклонный*** `***Bold and Italic***`
- ~~зачёркнутый шрифт~~ `~~Scratch Me~~`
- <u>шрифт с нижним подчёркиванием<u> `<u> underlined </u>`

Markdown позволяет выделять целые предложения или отдельные слова, комбинируя полужирный и курсивный шрифт.

## Вставка ссылок

Для вставки активных ссылок для прямого перехода используется следующий синтаксис: `[Заголовок ссылки](URL адрес)`.
URL адреса в квадратных скобках автоматически становятся ссылками.
```
[Страница кафедры информационной безопасности ВМК МГУ имени М.В. Ломоносова](https://ib.cs.msu.ru)
```
*Результат:*

[Страница кафедры информационной безопасности ВМК МГУ имени М.В. Ломоносова](https://ib.cs.msu.ru)

Есть ещё один вариант с использованием языка HTML (HyperText Markup Language — «язык гипертекстовой разметки»): 
```
<a href="https://www.3blue1brown.com">Animated Math</a>
```
*Результат:*

<a href="https://www.3blue1brown.com">Animated Math</a>

При использовании HTML могут возникать ошибки из-за использования служебных символов, например, пробела или скобок в примерах ниже:

1.
    ```
    [link](https://www.example.com/my great page)
    ```
    <br>
    
    *Результат:*
    
    [link](https://www.example.com/my great page)
    
    Исправленный код:
    ```
    [link](https://www.example.com/my%20great%20page)
    ```
    <br>
    
    *Результат:*
    
    [link](https://www.example.com/my%20great%20page)

2. Стоит писать `[a novel](https://en.wikipedia.org/wiki/The_Milagro_Beanfield_War_%28novel%29)`

   
   вместо `[a novel](https://en.wikipedia.org/wiki/The_Milagro_Beanfield_War_(novel))`

## Применение HTML-языка в Markdown

К сожалению, не все платформы поддерживают использование HTML-языка, но приведём полезные шаблоны для более удобного отображения информации:

|Задача |	Markdown	| HTML |
|:------|:--------:|:--------:|
|Курсив	 |`*текст*`	|`<em>текст</em>`|
|Жирный	|`**текст**`	| `<strong>текст</strong>` |
|Ссылка	| `[ссылка](url)` |	`<a href="url">ссылка</a>` |
|Изображение |	`![текст](img.jpg)` |	`<img src="img.jpg" alt="текст">` |
|Цвет текста |	нет |	`<span style="color:red;">текст</span>` |
|Управление размерами | нет |	`<img width="200">`|
|Подстрочный текст | нет |`<sub>subscript text</sub>` |
|Надстрочный текст | нет |`<sup>superscript text</sup> ` |

Причина, по которой не все Markdown-редакторы поддерживают HTML, заключается в безопасности: некоторые HTML-теги могут быть использованы злоумышленниками для внедрения вредоносного кода, особенно в веб-приложениях (например, `<script>`).

Обратим отдельно внимание на вставку и редактирование изображений посредством HTML, такой способ более удобный, так как поддерживает изменение размера изображения, рамку и цвета:

```
<img src="изображение.jpg" alt="Описание" width="300" height="200" style="border: 2px solid red;">
```

Отметим, что хотя альтернативный текст (`alt="Описание"`) не отображается, но позволяет решать ряд достаточно важных задач:
1. Alt-текст помогает сохранить смысл даже без визуального контента, что бывает важно в документации, технических статьях, обучающих материалах.
2. Поисковые системы (Google, Яндекс и др.) анализируют alt-текст, чтобы понять, что изображено на картинке. Это помогает странице лучше индексироваться и отображаться в результатах поиска.
3. Если изображение не загрузилось, браузер отобразит вместо него альтернативный текст.
4. Для людей с нарушением зрения, которые используют экранные дикторы, alt-текст зачитывается вслух, чтобы пользователь понимал, что изображено.

Язык HTML позволяет привязать подпись к изображению и сделать вывести на экран более лаконичным способом, для этого воспользуйтесь следующими шаблонами:
```html
<figure>
  <img src="https://example.com/image.jpg" alt="Описание изображения" width="400">
  <figcaption>Рис. 1: Подпись к изображению</figcaption>
</figure>
```

С центрированием изображения и подписи:

```html
<p align="center">
  <img src="https://example.com/image.jpg" alt="Описание" width="300">
</p>
<p align="center"><em>Рис. 1: Центрированная подпись</em></p>
```


Следующие HTML-теги поддерживаются почти всегда:
- `<br>` — перенос строки
- `<img>` — вставка изображений
- `<a>` — гиперссылки
- `<strong>`, `<em>` — выделение текста
- `<hr>` — горизонтальная линия
- `<code>`, `<pre>` — код
  
В то время как
- `<div>`, `<table>`, `<section>`, `<article>` — отключены в некоторых редакторах
- `<script>`, `<iframe>`, `<style>` — часто блокируются из соображений безопасности

Рекомендации:
1. При использовании блочных HTML-элементов (`<div>`, `<table>`, `<pre>`, `<p>` и т.д.), всегда отделяйте их от остального текста пустой строкой сверху и снизу. Это помогает избежать конфликтов между HTML и Markdown-разметкой.
2. Не используйте отступы (пробелы или табуляцию) перед HTML-тегами.
3. Внутри HTML-блока Markdown-синтаксис не работает.

## Таблицы 

Самая простая таблица в Markdown имеет следующий синтаксис и выравнивается по центру страницы автоматически:
```
| Syntax      | Description |
| ----------- | ----------- |
| Header      | Title       |
| Paragraph   | Text        |
```

*Результат:*
| Syntax      | Description |
| ----------- | ----------- |
| Header      | Title       |
| Paragraph   | Text        |

Следующие последовательности символов отвечают за
- `|` отделение ячеек 
- `----` горизонтальную черту
- `:--- ` выравнивание по левому краю
- `:----: ` выравнивание по центру
- ` ---:  ` выравнивание по правому краю

Пример таблицы с использованием выравнивания текста по столбцам:
```
| Syntax      | Description | Test Text     |
| :---        |    :----:   |          ---: |
| Header      | Title       | Here's this   |
| Paragraph   | Text        | And more      |
```

*Результат:*

| Syntax      | Description | Test Text     |
| :---        |    :----:   |          ---: |
| Header      | Title       | Here's this   |
| Paragraph   | Text        | And more      |


Обращаем внимание, что синтаксис Markdown не поддерживает объединение ячеек. Для более сложных конструкций необходимо пользоваться HTML.
