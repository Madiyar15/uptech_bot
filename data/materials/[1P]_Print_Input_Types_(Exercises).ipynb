# Основные определения

**Машинный код** — система команд конкретной вычислительной машины, которая интерпретируется непосредственно микропроцессором или микропрограммами данной вычислительной машины.

**Компьютерная программа** — список команд (инструкций) для компьютера. Команды могут быть любыми, например:
- считать информацию с клавиатуры;
- произвести арифметические вычисления (+, −, *, /);
- вывести информацию на экран.

**Язык программирования** — набор определенных правил, по которым компьютер может понимать команды (инструкции) и выполнять их.

**Программный код** - текст программы на любом языке программирования.


Языки программирования бывают **компилируемые** и **интерпретируемые**. Если программа написана на компилируемом языке (C, C++, Pascal), то перед выполнением её нужно полностью проверить на наличие синтаксических ошибок и уже после этого перевести в понятную для компьютера форму — **машинный код**. Это делает специальная программа, которая называется **компилятором**. Если программа написана на интерпретируемом языке (Python, PHP, Ruby), она не переводится в машинный код целиком. Вместо этого специальная программа, которая называется интерпретатором, идет по коду, анализирует его и выполняет каждую отдельную команду.

Существуют языки программирования, которые совмещают оба подхода (C#, Java). В таких языках код исходной программы сначала компилируется в промежуточный код (байт-код), а уже потом, во время выполнения, переводится в машинный код.

**Компиляция** - это перевод программы из одного языка в другой, обычно из высокоуровневого в более низкоуровневое представление (машинный код, объектный код, байткод или IR (Intermediate Representation)), часто с анализом и оптимизациями. Результатом обычно является артефакт (исполняемый файл или объектные модули), который затем запускается напрямую или после линковки/в виртуальной машине.

**Интерпретация** – это способ выполнения программы, при котором интерпретатор во время запуска анализирует исходный код или его промежуточное представление (например, байткод) и немедленно исполняет соответствующие операции, обычно без предварительной компиляции в отдельный исполняемый файл.

Python сочетает оба подхода. При запуске программы код сначала компилируется в байт, а затем интерпретируется. Другими словами, компиляция - это процесс, который происходит один раз, а затем программа запускается.

Интерпретация происходит постепенно. Часть кода интерпретируется и запускается, а затем еще небольшая часть кода интерпретируется и запускается. Интерпретируемые языки нуждаются в интерпретаторе, чтобы постоянно выполнять небольшие фрагменты кода.

| ``a % b``    | Modulus        | Целочисленный от деления``a`` на ``b``     |
| ``a ** b``   | Exponentiation | ``a`` возведённая в степень числа ``b``                     |
| ``-a``       | Negation       | "Отрицание"``a``                                  |

Иногда мы хотим изменить порядок действий при вычислении некоторого выражения. Тут нам на помощь приходят скобки. В Python можно использовать круглые скобки для установки приоритета операций. Квадратные скобки `[]` и фигурные скобки `{}` зарезервированы для других целей, с которыми мы скоро познакомимся.

````python
hat_height_cm = 25
my_height_cm = 190
# How tall am I, in meters, when wearing my hat?
total_height_meters = (hat_height_cm + my_height_cm) / 100
print("Height in meters =", total_height_meters)
````

Более общие правила определения приоритетов операций такие:

1. Выполняются возведения в степень справа налево, то есть 3 ** 3 ** 3 это $3^{(3^3)}$.
2. Выполняются унарные минусы (отрицания).
3. Выполняются умножения и деления слева направо. Операции умножения и деления имеют одинаковый приоритет.
4. Выполняются сложения и вычитания слева направо. Операции сложения и вычитания имеют одинаковый приоритет.

Основные операции над строками:
- A + B — конкатенация;
- A * n — повторение n раз, значение n должно быть целого типа.

Перед тем как мы сможем продвинуться дальше, нам следует обсудить именования переменных в Python.

Имена переменных обязаны начинаться с символа-буквы, причём имена переменных принято начинать со строчной буквы, а имена классов — с заглавной (с классами мы подробнее познакомимся позже).

Кроме того, существует ряд ключевых слов Python, которые нельзя использовать в качестве имен переменных. Эти ключевые слова:
```
    and, as, assert, break, class, continue, def, del, elif, else, except,
    exec, finally, for, from, global, if, import, in, is, lambda, not, or,
    pass, print, raise, return, try, while, with, yield
```

**Замечание:** Помните о ключевом слове `lambda`, которое вполне может быть естественным именем переменной в решающей научную задачу программе. Поскольку это название зарезервировано под ключевое слово, его нельзя использовать в качестве имени переменной.

Оператор присваивания сообщает переменной то или иное значение независимо от того, была ли эта переменная введена раньше. Вы можете менять значение переменной, записав еще один оператор присваивания. ## Приведение типов

В языке Python определены функции преобразования типов, которые позволяют удобно преобразовывать одни типы данных в другие.

### Неявное приведение типов

При неявном приведении типов в Python, интерпретатор Python автоматически конвертирует один тип данных в другой, не требуя каких-либо действий от пользователя. Давайте рассмотрим эту процедуру на небольшом примере.

````python
x = 10

print("x is of type:", type(x))

y = 10.6
print("y is of type:", type(y))

z = x + y

print(z)
print("z is of type:", type(z))
````

В данном блоке кода:

1. Мы объявили целочисленную переменную `x` и вещественную переменную `y`
1. Сложили `x`и `y`, поместив результат в переменную `z`
1. В результатах исполнения ячейки кода мы заметили, что объект `x+y` имеет вещественный тип

Как мы можем заметить, тип переменной `z` был автоматически выбран в качестве "float", при том что она была получена при операции сложения объектов типа "float" и "int":

$$\text{"int"} \, + \, \text{"float"} \, \rightarrow \, \text{"float"}.$$

При неявном преобразовании типов работает общее правило (т.н. type promotion), согласно которому результирующий тип данных должен является "расширением" базового и допускать преобразование без потери информации. Именно поэтому переменная `z` получилась вещественной, а не целочисленной.

Неявное приведение типов не всегда возможно. Если мы попытаемся сложить строку и число, интерпретатор Python оповестит нас об ошибке:

````python
x = 10

print("x is of type:",type(x))

y = 'Ten point six'
print("y is of type:",type(y))

z = x + y

print(z)
print("z is of type:",type(z))
````

В данном блоке кода:

1. Мы объявили целочисленную переменную `x` и строку `y`
2. Попытались вычислить объект `x+y` и присвоить результат `z`
3. Строковая переменная `y` не была автоматически преобразована в число и мы получили `TypeError` в качестве результата работы ячейки

### Явное приведение типов

При явном приведении типов, как это следует из названия, от нас потребуется вручную выбрать тип данных, к которому мы хотим осуществить преобразование. В отличие от неявного приведения типов, явное преобразование может проходить с потерей информации. Рассмотрим несколько примеров:

1. `int()`:  Данная функция осуществляет попытку преобразования произвольного объекта в целое число.
2. `float()`: Данная функция осуществляет попытку преобразования произвольного объекта в число с плавающей запятой

````python
# Python code to demonstrate Type conversion using int(), float()

# initializing string
s = "42"

# printing string converting to int
c = int(s)
print ("After converting to integer : ", end="")
print (c)

# printing string converting to float
e = float(s)
print ("After converting to float : ", end="")
print (e)
````

3. `ord()` : используется для преобразования символа (character) в целое число
4. `hex()` : используется для преобразования целого в строку с его шестнадцатиричной записью
5. `oct()` : используется для преобразования целого в строку с его восьмеричной записью

````python
# Python code to demonstrate Type conversion
# using  ord(), hex(), oct()

# initializing integer
s = '4'

# printing character converting to integer
c = ord(s)
print("After converting character to integer : ",end="")
print(c)

# printing integer converting to hexadecimal string
c = hex(42)
print("After converting 42 to hexadecimal string : ",end="")
print(c)

# printing integer converting to octal string
c = oct(42)
print("After converting 42 to octal string : ",end="")
print(c)
````

Вернёмся к рассмотренному выше примеру, где мы пытались сложить строку и число. Попробуем исправить `TypeError`, явно приведя строку `'Ten point six'`к типу `float`:

````python
x = 10
print("x is of type:",type(x))

y = 'Ten point six'
print("y is of type:",type(y))

z = x + float(y)

print(z)
print("z is of type:",type(z))
````

Явное привидение типов в данном примере нам не помогло получить желаемый результат. Чтобы преобразовать строку `'Ten point six'`в вещественное число `10.6` придётся провести гораздо больше работы.

Тем не менее, явное приведение типов может нам помочь преобразовать строковую запись числа и само число: $\text{'10.6'} \rightarrow \text{10.6}$

````python
x = 10
print("x is of type:",type(x))

y = '10.6'
print("y is of type:",type(y))

z = x + float(y)

print(z)
print("z is of type:",type(z))
````

# Ввод данных с клавиатуры

Программы, написанные на языке Python, могут считывать данные, а затем их использовать.

Для считывания данных в языке Python используется функция `input()` (всегда пишется с круглыми скобками), которая  работает так: когда программа доходит до места, где есть `input()`, она ждёт, пока пользователь введет текст с клавиатуры (ввод завершается нажатием клавиши Enter). Введенная строка подставляется на место `input()`.



````python
print('Как тебя зовут?')     # вывод текста
print('Привет,', input(), '!')    # ввод текста и вывод текста
````

Чтобы записать в значение переменной то, что должно быть по условию задачи введено с клавиатуры, достаточно просто присвоить переменной саму функцию `input()`:
```
a = input()
b = input()
```
В приведённом блоке кода функция `input()` считывает строку с клавиатуры и возвращает значение считанной строки, которое сразу же присваивается переменым `a` и `b`.


Можно объединить считывание строк и преобразование типов, если вызывать функцию `int` для того значения, которое вернет функция `input`:

# Рекомендации к оформлению кода на языке Python: PEP 8 – Style Guide for Python Code

PEP8 — это руководство по стилю написания кода на Python: как именовать переменные и функции, оформлять отступы, строки, импорты и так далее. Его цель — сделать код единообразным и читаемым. PEP 8 периодически обновляется как «живой» документ, актуальная редакция PEP 8 всегда доступна на [официальном сайте](https://peps.python.org/pep-0008/).

**Рекомендация 1.** Не используйте пробел перед открывающей скобкой с аргументами команды `print()`.

Правильно:
```
print('Follow PEP8!')
```
         
Неправильно:
```
print ('Follow PEP8!')
```
     
**Рекомендация 2.** Используйте пробел после запятой, при перечислении аргументов в команде `print()`.

Правильно:
```
print('PEP8', 'Rocks!')
```
               
Неправильно:
```
print('PEP8','Rocks!')
```
     
**Рекомендация 3.** Не отделяйте пробелами оператор присваивания `=` при определении параметра по умолчанию.

Правильно:
```
print('My name', 'is', 'Python', sep='**', end='+')
```
              
Неправильно:
```
print('My name', 'is', 'Python', sep = '**', end = '+')
```
     
**Рекомендация 4.** Отделяйте пробелами оператор присваивания `=` при определении переменных.

Правильно:
```
a = 10
b = 20
```
                  
Неправильно:
```
a=10
b=20
```

**Рекомендация 5.** Согласно стандарту PEP 8, комментарии должны отделяться хотя бы двумя пробелами от кода. Они должны начинаться с символа `#` и одного пробела.

Правильно:
```
# Далее будет напечатан текст
print('Comments in Python')  # Печать текста с помощью команды print.
```
                  
Неправильно:
```
#Далее будет напечатан текст

```
