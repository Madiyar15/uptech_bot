# f-строки

f-строки («formatted string») - это пятый способ форматирования строк в Python,  который работает быстрее, чем другие способы форматирования. Они также улучшают читаемость кода, и задаются с помощью литерала `f` перед кавычками.

f-строки просто берут значения переменных, которые есть в текущей области видимости, и подставляют их в строку. В самой строке Вам всего лишь нужно указать имя этой переменной в фигурных скобках.

Источник: [F-strings (Shultas Education)](https://shultais.education/blog/python-f-strings)

````python
name = "Дмитрий"
age = 25
print(f"Меня зовут {name}. Мне {age} лет.")
````

Отличие от предыдущих способов форматирования:

````python
integer = 42
string = 'FORTY_TWO'

print('string number %s, or simply %d' % (string, integer))
print('string number {}, or simply {}'.format(string, integer))
print(f'string number {string}, or simply {integer}')
````

Источники:
[F-strings или как сделать код чуть более быстрым и читаемым](https://habr.com/ru/articles/462179/)

f-строки также поддерживают расширенное форматирование чисел:

````python
from math import pi
print(f"Значение числа pi: {pi:.2f}")
````

С помощью f-строк можно форматировать дату без вызова метода strftime():


````python
from datetime import datetime as dt
now = dt.now()
print(f"Текущее время {now:%d.%m.%Y %H:%M}")
````

f-строки позволяют вычислять значения вырадений прямо в самой строке:

````python
x = 10
y = 5
print(f"{x} x {y} / 2 = {x * y / 2}")
````

При помощи f-строк можно обращаться к элементам списка прямо по индексу:

````python
planets = ["Меркурий", "Венера", "Земля", "Марс"]
print(f"Мы живём на планете {planets[2]}!")
````

f-строки позволяют обращаться к элементам словаря по ключу, причём как к целочисленному, так и строчному типу ключа:

````python
planet = {"name": "Земля", "radius": 6378000}
print(f"Планета {planet['name']}. Радиус {planet['radius']/1000} км.")
````

````python
digits = {0: 'ноль', 'one': 'один'}
print(f"0 - {digits[0]}, 1 - {digits['one']}")
````

Можно сразу применять методы к объектам:

````python
name = "Дмитрий"
print(f"Имя: {name.upper()}")
````

И вызывать значения функций:

````python
print(f"13 / 3 = {round(13/3)}")
````

Начиная с Python версии 3.12, f-строки поддерживают так называемое переиспользование кавычек (quote reuse), когда один вид кавычек применяется и для оформления самой строки, и для работы внутри фигурных скобок. Иными словами, теперь вы можете писать так:

````python
planet = {"name": "Земля", "radius": 6378000}
print(f"Планета {planet["name"]}. Радиус {planet["radius"]/1000} км.")
````

# Генераторы списков

Генераторы списков строятся по следующему шаблону:
```
[выражение for переменная in последовательность]
```
                  
Здесь используются обозначения:

- выражение — то, что вычисляется на каждой итерации и добавляется в список;
- переменная — имя, которое принимает значение текущего элемента из последовательности;
- последовательность — объект, по которому осуществляется итерация (например, список, диапазон range или другой итерируемый объект).
Теперь разберем несколько примеров.

Пример. Генератор списка, состоящего из 7 нулей:
```
zeros = [0 for _ in range(7)]
print(zeros)
[0, 0, 0, 0, 0, 0, 0]
```

```

```

print(numbers)

print(even_numbers)

```

```

# Функции в Python

**Функция** — отдельная функционально независимая часть программы, решающая определённую задачу.

**Заголовок функции** — первая строка определения функции.

**Тело функции** — программный код, содержащийся в функции.

**Аргумент** – любая порция данных, которая передается в функцию, когда функция вызывается.

**Параметр** – переменная, которая получает аргумент, переданный в функцию.


Итак, вот что следует знать об аргументах функций:

- Порядок передачи функциям позиционных аргументов.
- Порядок передачи функциям именованных аргументов.
- Назначение значений аргументов, применяемых по умолчанию.
- Организация обработки наборов аргументов переменной длины.
- Распаковка аргументов.
- Использование аргументов, которые можно передавать только по имени (keyword-only).

## Определение функций

Python предоставляет широкий набор встроенных функций, но нам обязательно понадобятся собственные функции. Давайте рассмотрим простой пример:

````python
# Define the function
def add_three(input_var):
    output_var = input_var + 3
    return output_var
````

Каждая функция состоит из двух частей: **зоголовка** (header) and **тела функции** (body).


![](https://edunet.kea.su/repo/python-bootcamp/src/L01/func_00.png)

[Источник изображения](https://www.kaggle.com/code/alexisbcook/functions?scriptVersionId=97654315&cellId=3)

### Header
Заголовок функции определяет её название и какими аргументами она обладает.
- Каждый заголовок функции начинается с ключевого слова `def`, которая указывает python на начало объявления функции.
- В нашем примере названием функции является `add_three`.  
- В нашем примере аргументом функции является `input_var`. *Аргументом* функции называют переменную, которая будет использоваться в качестве входного значения для данной функции. Аргументы всегда перечисляются в круглых скобках, которые следуют сразу после объявления названия функции. Отметим, что функция может вовсе не иметь аргументов или наоборот иметь несколько, перечисленных через запятую.
- При объявлении любой функции заголовок обязан заканчиваться символом двоеточия `:`

### Body
**Тело** (body) функции определяет всю работу, которую должна делать функция.
- Каждая строка кода в теле функции должная иметь отступ в 4 пробела с начала строки. Отступы от начала строки организуют код на Python в логические блоки (как фигурные скобки в некоторых других языках программирования).
- Функция выполняет свою работу путём последовательно запуска строк с нужным отступом сверху внизу
    - Она берёт аргументы в качестве входных значений (`input_var` в нашем примере)
    - Выполняет с этими переменными действия, реализуя некоторый алгоритм
    - Когда выполнение доходит до строки с ключевым словом **return**, функция назначает результатом свой работы переменную (или несколько переменных), которые следуют через пробел от данного ключевого слова --- выполнение функции завершается


Описанная конструкция языка Python только определяет функцию, но не вызывает её.

## Как запустить (вызвать) функцию

When we run a function, it can also be referred to as "calling" the function.

В ячейке ниже мы запускаем объявленную нами функцию со значением `10` в качесвте аргумента. Мы определим новую переменную `new_number`, которая присваивает себе значение, которое возвращает функция.

````python
# Run the function with 10 as input
new_number = add_three(10)

# Check that the value is 13, as expected
print(new_number)
````

Или более детально:

  - Сперва вычисляется `output_var = input_var + 3`, из-за чего `output_var = 13`.  
  - Последняя строка кода в теле функции содержит выражение return, которое возвращает значение переменной `output_var`, которое равно`13`.
- Выполняя присваивание `new_number = add_three(10)`, мы наделяем переменную `new_number` значением 13.

## Аргументы функции

Функция может принимать произвольное количество аргументов или не принимать их вовсе. Также распространены функции с произвольным числом аргументов, функции с позиционными и именованными аргументами, обязательными и необязательными.

**Позиционными** называются аргументы, передаваемые без указания имён. Их значения сопоставляется с параметрами функции по позиции.

**Именованными** называются аргументы, передаваемые с указанием имени параметра. Порядок упоминания именованных аргументов не имеет значения.

Практикуется подход: если функция принимает больше трёх аргументов, нужно хотя бы часть из них указать по имени. Особенно важно именовать значения аргументов, когда они относятся к одному типу, ведь без имен очень трудно понять, что делает функция с подобным вызовом.

## Параметры функции

Важно понимать разницу между аргументами и параметрами функции:

**Аргумент** – любая порция данных, которая передается в функцию, когда функция вызывается.

**Параметр** – переменная, которая получает аргумент, переданный в функцию.

Контрольный вопрос:

Дана функция `draw_box()`:
  ```python
  def draw_box(height, width):
      for i in range(height):
          print('*' * width)
  height = 10
  draw_box(height, 9)
  ```
определить аргументы и параметры.

Когда аргумент передается в функцию, параметрическая переменная функции будет ссылаться на значение этого аргумента. Однако любые изменения, которые вносятся в параметрическую переменную, не будут влиять на аргумент.

Важно помнить, что
- параметры могут быть позиционными и именованными
- именованные должны всегда идти после позиционных
- именованные при вызове можно помещать в любом порядке за позиционными
- именованные подразумевают задание значений по умолчанию

Написать две функции: для вычисления факториала и для вычисления числа сочетаний. Вычислить $C_5^3$.

````python
def factorial(x):
    pr = 1
    for i in range(2, x + 1):
        pr *= i
    return pr

def combinations(n, k):
    return factorial(n) / (factorial(k) * factorial(n - k))

print(combinations(5, 3))
````

## Локальные и глобальные переменные

**Локальными** называются переменные, объявленные внутри функции и доступные только ей самой. Программный код за пределами функции к ним доступа не имеет.

**Область видимости переменной** – часть программы, в которой можно к ней обращаться (та функция, где она создана). Переменная видима программному коду только в области её видимости. Никакая инструкция за пределами функции не может обращаться к такой переменной.

**Область видимости параметрической переменной** — функция, в которой этот параметр используется. К параметрической переменной имеет доступ весь программный код этой функции.

**Глобальными** называются переменные, объявленные в основной программе и доступные как программе, так и всем ее функциям.

**Глобальная константа** – глобальное имя, ссылающееся на неизменное значение.

Если внутри функции нужно присвоить значение глобальной переменной, её нужно объявить с помощью ключевого слова `global`.

 Значение по умолчанию для параметра создается единожды при определении функции (обычно при загрузке модуля) и становится атрибутом (свойством) функции. Поэтому, если значение по умолчанию изменяемый объект, то его изменение повлияет на каждый следующий вызов функции.

Чтобы посмотреть значения по умолчанию, можно использовать атрибут `__defaults__`.

````python
def append(element, seq=[]):
    seq.append(element)
    return seq

print('Значение по умолчанию', append.__defaults__)
print(append(10))
print('Значение по умолчанию', append.__defaults__)
print(append(5))
print('Значение по умолчанию', append.__defaults__)
print(append(1))
print('Значение по умолчанию', append.__defaults__)
````

Для решения проблемы можно использовать константу `None` в качестве значения параметра по умолчанию, а в теле функции устанавливать нужное значение:

````python
def append(element, seq=None):
    if seq is None:
        seq = []
    seq.append(element)
    return seq

print('Значение по умолчанию', append.__defaults__)
print(append(10))
print('Значение по умолчанию', append.__defaults__)
print(append(5))
print('Значение по умолчанию', append.__defaults__)
print(append(1))
print('Значение по умолчанию', append.__defaults__)
````

## Множественные переменные

Организация обработки наборов аргументов переменной длины реализована при помощи конструкций `*args` и `**kwargs`. Когда эти конструкции используются при объявлении функции, мы ожидаем, что при вызове функции наборы аргументов произвольной длины будут представлены в виде параметров `args` и `kwargs`:
- при применении конструкции `*args` в параметр `args` попадают позиционные аргументы, представляемые в виде кортежа;
- при применении `**kwargs` в `kwargs` попадают именованные аргументы, представленные в виде словаря.

`*args` — сокращение от «arguments» (аргументы), а `**kwargs` — сокращение от «keyword arguments» (именованные аргументы).

В параметре `args` хранится кортеж, содержащий то, что передано функции при её вызове:

````python
def foo(*args):
    print(args)

a, b, c = 1, 2, 3

foo(a, b, c)                # Выводит (1, 2, 3)
foo(a, b)                   # Выводит (1, 2)
foo(a)                      # Выводит (1)
foo(b, c)                   # Выводит (2, 3)
````

В параметре `kwargs` хранится словарь, состоящий из пар ключ-значение и представляющий именованные аргументы, переданные функции при вызове:

````python
def foo(**kwargs):
    print(kwargs)

foo(a=1, b=2, c=3)        # Выводит {'a': 1, 'b': 2, 'c': 3}
foo(a=1, b=2)             # Выводит {'a': 1, 'b': 2}
foo(a=1)                  # Выводит {'a': 1}
foo(b=2, c=3)             # Выводит {'b': 2, 'c': 3}
````

Надо отметить, что функции, рассчитанной на приём позиционных аргументов, нельзя передавать именованные аргументы (и наоборот):

````python
def foo(*args):
    print(args)

foo(a=1, b=2, c=3)
````

````python
def foo(**kwargs):
    print(kwargs)

a, b, c = 1, 2, 3
foo(a, b, c)
````

## Анонимные функции, инструкция `lambda`


Помимо стандартного определения функции, состоящего из ее заголовка с ключевым словом def и блока кода – тела функции, в Python можно создавать короткие однострочные функции с использованием оператора `lambda`. Это анонимные функции или лямбда-функции.

Общий формат определения анонимной функции: `lambda список_параметров: выражение`.

Тут `список_параметров` – список параметров через запятую, `выражение` – значение, либо код, дающий значение.

Анонимные функции могут содержать лишь одно выражение, но и выполняются они быстрее. Анонимные функции создаются с помощью инструкции `lambda`.

````python
def standard_function(x):            #  стандартное объявление функции
    return x*2

lambda_function = lambda x: x*2      #  объявление анонимной функции

print(standard_function(7))
print(lambda_function(7))
````
